```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import regex as re
```

```python
table_data=[]
ranks=[]
modified=[]
revenue=[]
net_operating_income=[]
profit=[]
assets=[]
net_worth=[]
equity_dividend=[]
employee_cost=[]
company_name=[]
year=[]

for i in range(2010,2021):
    year.append(i)
    modified=[]
    r=[]
    print(i)
    if i==2011:
        url = 'https://www.fortuneindia.com/fortune-500/company-list/g-a-i-l-india?year='+str(i)
        page = requests.get(url)
        soup = BeautifulSoup(page.text,'html.parser')
        a=soup.findAll('div',{'class':'sl-num'})
        table=soup.findAll('div',{'class':'company-parameters'})
        for i in table:
            table_data.append(i.text.replace('\n',' '))
        for i in a:
            r.append(i.text.replace('\n',''))
        for i in r:
            ranks.append(i[0])
    else:
        url = 'https://www.fortuneindia.com/fortune-500/company-list/gail-india?year='+str(i)
        page = requests.get(url)
        soup = BeautifulSoup(page.text,'html.parser')
        a=soup.findAll('div',{'class':'sl-num'})
        table=soup.findAll('div',{'class':'company-parameters'})
        for i in table:
            table_data.append(i.text.replace('\n',' '))
        for i in a:
            r.append(i.text.replace('\n',''))
        for i in r:
            ranks.append(i[:3])
        
            
    company_name.append('GAIL')
   



for i in table_data:
    modified.append(i[4:].split())
for i in modified:
    revenue.append(i[6])
for i in modified:
    net_operating_income.append(i[11])
for i in modified:
    profit.append(i[14])
for i in modified:
    assets.append(i[17])
for i in modified:
    net_worth.append(i[21])
for i in modified:
    equity_dividend.append(i[25])
for i in modified:
    employee_cost.append(i[29])
    
gail_dic={'Company_name':company_name,'Ranks':ranks,'year':year,
              'Revenue':revenue,
             'Net operating income':net_operating_income,'profit':profit,
             'Assets':assets,'Net worth':net_worth,'equity dividend':equity_dividend,
             'employee cost':employee_cost}

```
